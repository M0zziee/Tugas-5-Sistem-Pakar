```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Generate universe of discourse for temperature
suhu = np.arange(35, 41, 0.1)

# Generate fuzzy membership functions
rendah = fuzz.trimf(suhu, [35, 35, 36.5])
sedang = fuzz.trimf(suhu, [36, 37, 38])
tinggi = fuzz.trimf(suhu, [37.5, 39, 40])

# Plot the membership functions
plt.figure()
plt.plot(suhu, rendah, 'b', label='Rendah')
plt.plot(suhu, sedang, 'g', label='Sedang')
plt.plot(suhu, tinggi, 'r', label='Tinggi')
plt.title('Fuzzy Set Suhu Tubuh')
plt.xlabel('Suhu (derajat C)')
plt.ylabel('Derajat Keanggotaan')
plt.legend()
plt.grid(True)
plt.show()

# Test with a specific temperature value
input_suhu = 35.2
suhu_rendah = fuzz.interp_membership(suhu, rendah, input_suhu)
suhu_sedang = fuzz.interp_membership(suhu, sedang, input_suhu)
suhu_tinggi = fuzz.interp_membership(suhu, tinggi, input_suhu)

print(f"Derajat keanggotaan suhu {input_suhu} derajat C:")
print(f"- Rendah: {suhu_rendah:.2f}")
print(f"- Sedang: {suhu_sedang:.2f}")
print(f"- Tinggi: {suhu_tinggi:.2f}")
```

Fuzzy Klasifikasi Tingkat Air Dalam Tangki

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Universe of discourse
level = np.arange(0, 101, 1)

# Fuzzy sets
rendah = fuzz.trapmf(level, [0, 0, 20, 40])
sedang = fuzz.trimf(level, [30, 50, 70])
tinggi = fuzz.trapmf(level, [60, 80, 100, 100])

# Visualize fuzzy sets
plt.figure(figsize=(10, 6))
plt.plot(level, rendah, 'b', linewidth=1.5, label='Rendah')
plt.plot(level, sedang, 'g', linewidth=1.5, label='Sedang')
plt.plot(level, tinggi, 'r', linewidth=1.5, label='Tinggi')
plt.title('Fuzzy Level Air dalam Tanah')
plt.xlabel('Level Air (%)')
plt.ylabel('Derajat Keanggotaan')
plt.legend()
plt.grid(True)
plt.show()

# Calculate membership for a specific input
input_level = 25
level_rendah = fuzz.interp_membership(level, rendah, input_level)
level_sedang = fuzz.interp_membership(level, sedang, input_level)
level_tinggi = fuzz.interp_membership(level, tinggi, input_level)

print(f"Derajat keanggotaan level air {input_level}%:")
print(f"- Rendah: {level_rendah:.2f}")
print(f"- Sedang: {level_sedang:.2f}")
print(f"- Tinggi: {level_tinggi:.2f}")
```

Fuzzy Air Quality Index

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Universe of discourse
aqi = np.arange(0, 301, 1)

# Fuzzy sets for Air Quality Index
baik = fuzz.trapmf(aqi, [0, 0, 25, 50])
sedang = fuzz.trimf(aqi, [25, 75, 125])
tidak_sehat = fuzz.trimf(aqi, [100, 150, 200])
sangat_tidak_sehat = fuzz.trimf(aqi, [175, 225, 275])
berbahaya = fuzz.trapmf(aqi, [250, 275, 300, 300])

# Visualize fuzzy sets
plt.figure(figsize=(12, 6))
plt.plot(aqi, baik, 'g', linewidth=1.5, label='Baik')
plt.plot(aqi, sedang, 'y', linewidth=1.5, label='Sedang')
plt.plot(aqi, tidak_sehat, 'r', linewidth=1.5, label='Tidak Sehat')
plt.plot(aqi, sangat_tidak_sehat, 'r', linewidth=1.5, label='Sangat Tidak Sehat')
plt.plot(aqi, berbahaya, 'purple', linewidth=1.5, label='Berbahaya')
plt.title('Fuzzy Set Indeks Kualitas Udara (AQI)')
plt.xlabel('AQI')
plt.ylabel('Derajat Keanggotaan')
plt.legend()
plt.grid(True)
plt.show()

# Calculate membership for a specific input
input_aqi = 120
aqi_baik = fuzz.interp_membership(aqi, baik, input_aqi)
aqi_sedang = fuzz.interp_membership(aqi, sedang, input_aqi)
aqi_tidak_sehat = fuzz.interp_membership(aqi, tidak_sehat, input_aqi)
aqi_sangat_tidak_sehat = fuzz.interp_membership(aqi, sangat_tidak_sehat, input_aqi)
aqi_berbahaya = fuzz.interp_membership(aqi, berbahaya, input_aqi)

print(f"Derajat keanggotaan AQI {input_aqi}:")
print(f"- Baik: {aqi_baik:.2f}")
print(f"- Sedang: {aqi_sedang:.2f}")
print(f"- Tidak Sehat: {aqi_tidak_sehat:.2f}")
print(f"- Sangat Tidak Sehat: {aqi_sangat_tidak_sehat:.2f}")
print(f"- Berbahaya: {aqi_berbahaya:.2f}")
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

# Input variables
temperature = ctrl.Antecedent(np.arange(15, 41, 1), 'temperature')
humidity = ctrl.Antecedent(np.arange(20, 101, 1), 'humidity')

# Output variable
fan_speed = ctrl.Consequent(np.arange(0, 101, 1), 'fan_speed')

# Membership functions for temperature
temperature['dingin'] = fuzz.trapmf(temperature.universe, [15, 15, 20, 24])
temperature['nyaman'] = fuzz.trimf(temperature.universe, [21, 26, 30])
temperature['panas'] = fuzz.trapmf(temperature.universe, [28, 32, 40, 40])

# Membership functions for humidity
humidity['kering'] = fuzz.trapmf(humidity.universe, [20, 20, 30, 45])
humidity['normal'] = fuzz.trimf(humidity.universe, [35, 50, 65])
humidity['lembab'] = fuzz.trapmf(humidity.universe, [60, 75, 100, 100])

# Membership functions for fan speed
fan_speed['mati'] = fuzz.trimf(fan_speed.universe, [0, 0, 25])
fan_speed['rendah'] = fuzz.trimf(fan_speed.universe, [15, 30, 45])
fan_speed['sedang'] = fuzz.trimf(fan_speed.universe, [35, 50, 75])
fan_speed['tinggi'] = fuzz.trimf(fan_speed.universe, [65, 85, 95])
fan_speed['maksimal'] = fuzz.trimf(fan_speed.universe, [75, 100, 100])

# Visualize membership functions
temperature.view()
humidity.view()
fan_speed.view()

# Define fuzzy rules
rule1 = ctrl.Rule(temperature['dingin'] & humidity['kering'], fan_speed['mati'])
rule2 = ctrl.Rule(temperature['dingin'] & humidity['normal'], fan_speed['mati'])
rule3 = ctrl.Rule(temperature['dingin'] & humidity['lembab'], fan_speed['rendah'])
rule4 = ctrl.Rule(temperature['nyaman'] & humidity['kering'], fan_speed['rendah'])
rule5 = ctrl.Rule(temperature['nyaman'] & humidity['normal'], fan_speed['sedang'])
rule6 = ctrl.Rule(temperature['nyaman'] & humidity['lembab'], fan_speed['tinggi'])
rule7 = ctrl.Rule(temperature['panas'] & humidity['kering'], fan_speed['tinggi'])
rule8 = ctrl.Rule(temperature['panas'] & humidity['normal'], fan_speed['tinggi'])
rule9 = ctrl.Rule(temperature['panas'] & humidity['lembab'], fan_speed['maksimal'])

# Create control system
ac_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
sc = ctrl.ControlSystemSimulation(ac_ctrl)

# Test input
sc.input['temperature'] = 30
sc.input['humidity'] = 75

# Compute result
sc.compute()
print(f"Untuk suhu 30°C dan kelembaban 75%, kecepatan kipas AC: {sc.output['fan_speed']:.2f}%")

# Visualize result
fan_speed.view(sim=sc)
plt.show()
```

Certainty Factor (CF)

```python
# Diagnosis sederhana menggunakan CF

# Data gejala awal dan nilai CF
gejala = {
    "demam": 0.8,
    "batuk": 0.7,
    "pilek": 0.6
}

# Tambahkan 5 gejala baru
gejala.update({
    "sakit_kepala": 0.5,
    "lelah": 0.4,
    "sakit_tenggorokan": 0.6,
    "mual": 0.3,
    "nyeri_otot": 0.5
})

# Ubah nilai demam menjadi 0.2
gejala["demam"] = 0.2

# Fungsi menghitung CF gabungan
def hitung_cf(cf_values):
    cf_total = 0
    for i, cf in enumerate(cf_values):
        if i == 0:
            cf_total = cf
        else:
            if cf_total >= 0 and cf >= 0:
                cf_total = cf_total + cf * (1 - cf_total)
            elif cf_total < 0 and cf < 0:
                cf_total = cf_total + cf * (1 + cf_total)
            else:
                cf_total = (cf_total + cf) / (1 - min(abs(cf_total), abs(cf)))
    return round(cf_total, 3)

cf_akhir = hitung_cf(list(gejala.values()))
print("Nilai CF Gabungan:", cf_akhir)
```

Nilai demam diturunkan ke 0.2, sehingga pengaruhnya terhadap diagnosis berkurang drastis. Karena metode CF memperhitungkan kontribusi tiap gejala, penurunan kepercayaan pada gejala utama (seperti demam) menurunkan hasil akhir. Ini meniru proses berpikir dokter: semakin tidak yakin terhadap gejala, semakin lemah kesimpulan diagnosa.

Fuzzy Logic

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

# Variabel input suhu dan kelembaban
suhu = ctrl.Antecedent(np.arange(15, 41, 1), 'suhu')
kelembaban = ctrl.Antecedent(np.arange(0, 101, 1), 'kelembaban')
kipas = ctrl.Consequent(np.arange(0, 101, 1), 'kipas')

# Fungsi keanggotaan suhu
suhu['dingin'] = fuzz.trimf(suhu.universe, [15, 15, 22])
suhu['nyaman'] = fuzz.trimf(suhu.universe, [20, 25, 30])
suhu['panas'] = fuzz.trimf(suhu.universe, [28, 35, 40])

# Fungsi keanggotaan kelembaban
kelembaban['rendah'] = fuzz.trimf(kelembaban.universe, [0, 0, 50])
kelembaban['tinggi'] = fuzz.trimf(kelembaban.universe, [40, 100, 100])

# Fungsi keanggotaan kipas
kipas['lambat'] = fuzz.trimf(kipas.universe, [0, 0, 50])
kipas['sedang'] = fuzz.trimf(kipas.universe, [30, 50, 70])
kipas['cepat'] = fuzz.trimf(kipas.universe, [60, 100, 100])

# Aturan fuzzy
rule1 = ctrl.Rule(suhu['dingin'], kipas['lambat'])
rule2 = ctrl.Rule(suhu['nyaman'], kipas['sedang'])
rule3 = ctrl.Rule(suhu['panas'] & kelembaban['tinggi'], kipas['cepat'])
rule4 = ctrl.Rule(suhu['panas'] & kelembaban['rendah'], kipas['sedang'])

# Sistem kontrol fuzzy
kipas_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4])
simulasi = ctrl.ControlSystemSimulation(kipas_ctrl)

# Input suhu dan kelembaban
simulasi.input['suhu'] = 22
simulasi.input['kelembaban'] = 70

# Hitung output
simulasi.compute()
print("Kecepatan Kipas:", simulasi.output['kipas'])
```

Saat suhu 22°C, suhu masuk ke dua kategori yaitu "dingin" dan "nyaman" karena fungsi keanggotaan overlap.

Ini menunjukkan bagaimana fuzzy logic bisa menangani ambiguitas.

Dengan kelembaban 70%, dan suhu mendekati nyaman, maka sistem memilih kecepatan sedang sebagai hasil kompromi aturan.
